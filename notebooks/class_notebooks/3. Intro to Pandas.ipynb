# Setting an Index Column

In most time-series data, we want to view the other columns as functions of time. We can tell ``pandas`` to do this by using the [``set_index``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.set_index.html) function. Before we do this, let's look at ``df``.

```python
df.head(10)
```

```python
df.tail()
```

```python
df
```

Note that when we call ``set_index``, we use the optional parameter ``inplace=True`` to specify that we want ``df`` to be changed. If we didn't specify this, then the function would create a new ``DataFrame``, and leave ``df`` unchanged.

```python
df.set_index('Date', inplace=True)
```

```python
df
```

# Converting Date/Time types

Let's look at the ``Date`` column

```python
df.index
```

Note the line that says ``dtype: int64``. We know that it is a list of dates. We can tell ``pandas`` to parse the column as dates using the [``to_datetime``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html) method, and then assign that now-parsed ``Series`` to the ``Date`` column.

```python
df.index = pd.to_datetime(df.index, format="%Y%m")
df.index
```

```python
df.index = df.index.to_period('M')
```

```python
df.index
```

Similarly if we now access a single column, it is listed as a function of time.

```python
df['Lo 30']
```

# Accessing Columns as Series

The individual columns of a ``DataFrame`` are a data type called a [``Series``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html). We can list the column names of ``df`` using the [``columns``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.columns.html) property.

```python
df.columns
```

We can access an individual column using the ``[]`` operator and the name of column.

```python
df['Lo 30']
```

```python
df[['Lo 30']]
```

And we can use [``describe``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.describe.html) on a ``Series``.

```python
df[['Lo 30']].describe()
```

We can also create a ``DataFrame`` with a subset of the columns by using a list of column names in the ``[]``. For example.

```python
df[ ['Lo 30', 'Hi 30'] ]
```

# Accessing Rows

We can query an individual row by index using [``loc``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html).

```python
df.loc['2008-08']
```

Or we can access several rows using the ``:`` range syntax.

```python
 df.loc['2008':'2008']
```

# Plotting Data

We can use the [``plot``](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html) method to plot either the full ``DataFrame`` or a single series. After calling ``plot``, we will need to call ``matplotlib``'s [``show``](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.show.html) method.

```python
df.plot()
plt.show()
```

Note that the $x$ axis is the date. This is because we set the index column to be ``Date``.

We can also plot an individual column.

```python
df['Hi 30'].plot()
plt.show()
```
